```python
import numpy as np
import scipy.spatial as scisp
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
```

```python
# Time iteration
def step(t):
    e = 0
    for j in range(1,N-1):
        for k in range(1,N-1):
            # sum over the nn
            NN = np.sum([lat_B[j-1, k], lat_B[j+1,k], lat_B[j,k-1], lat_B[j,k+1]])
            # Compute the curvature
            Z = lat_B[j,k] - 1/2/D*NN
            # Check if node is unstable
            if np.abs(Z) > Z_c:
                lat_C[j,k] = lat_C[j,k] - (2*D/s)*Z_c
                # Distribution of energy to the neighbours
                # Create a nearest neighbours array
                nn = [[j-1,k], [j+1,k], [j,k-1], [j,k+1]]
                for neighbour in nn:
                    lat_C[neighbour[0], neighbour[1]] = lat_C[neighbour[0], neighbour[1]] + (1/s)*Z_c
                # nodal energy release
                g = (2*D/s)*(2*np.abs(Z)/Z_c-1)*Z_c**2
                # accumulate the energy release
                e = e + g
    #Check for avalanche
    if e>0:
        print('avalanche', 'time={}'.format(t))
        # iterate over lattice
        for j in range(1,N-1):
            for k in range(1,N-1):
                lat_B[j,k] = lat_B[j,k] + lat_C[j,k]
                lat_C[j,k] = 0
    # if no avalanches, then add energy
    else:
#         print('not avalanching')
        [randx,randy] = [np.random.randint(1,high = N-1), np.random.randint(1, high = N-1)]
        dB = np.random.random()*np.abs(sigma1-sigma2)+sigma1
        lat_B[randx, randy] += dB
    if not t%100:
        im = plt.imshow(lat_B, animated=True)
        mat_array.append([im])
    # Save the avalanche energy
    er.append(e)
    # save the lattice energy
    el.append(np.sum(lat_B**2))
    
```

```python
# Make matrix array for movie
fig = plt.figure()
mat_array = []
# Number of dimensions
D=2
# width of the lattice
N = 15
# Useful constant
s = 2*D+1
# Treshold
Z_c = 5
# Make the lattice for magnetic and work
lat_B = np.zeros((N,N))
lat_C = np.zeros((N,N))
# Time length of the iteration
t = int(5e3)
# Define energy range (average cant be 0)
sigma1, sigma2 = -0.2,0.8

# Make lists for energy plots
#Energy released
er = []
e0 = 2*D/s*Z_c**2
# Lattice energy
el = []

for i in range(t):
    step(i)
ani = animation.ArtistAnimation(fig, mat_array, interval=200, blit=True,
                                repeat_delay=0)
plt.show()
```

```python
plt.figure()
ax1 = plt.subplot(211)
plt.plot(np.arange(t), np.array(er)/e0)

# share x only
ax2 = plt.subplot(212, sharex=ax1)
plt.plot(np.arange(t), np.array(el)/e0)

```
